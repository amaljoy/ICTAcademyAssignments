```python
#Initial setup

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Question 1-Reading the dataset to python environment.

data = pd.read_excel("D:\Data Science and Analytics\Assignment 1\iris.xls")
```

```python
data =pd.DataFrame(data)
data
```

```python
#Question 2 - Displaying the columns in the dataset

for columns in data.columns:
    print(columns)
```

```python
#Question 3 - Calculating the mean of each column of the dataset

data.mean(axis=0 , skipna = True)
```

```python
#Question 4 - Checking for the null values present in the dataset

haveNull=data.isnull().values.any()
if (haveNull):
    print("There are null values")
else:
    print("There are no null values")
```

```python
#Question 5 - Plotting three meaning full visualizations

#Plot1- Sepal Width Vs Classification
plt.scatter(data['Classification'],data['SW'])
plt.title("Sepal Width Vs Classification", fontsize=20,pad=20)
plt.xlabel("Classification",fontsize=15,labelpad=10)
plt.ylabel("Sepal Width",fontsize=15,labelpad=10)
plt.show()
#Here we can see, Iris-setosa have larger Sepal width and Iris -versicolor have smaller Sepal width
```

```python
#Plot2 - Sepal Length of each species

data1=data.loc[data['Classification']=='Iris-setosa']
data2=data.loc[data['Classification']=='Iris-versicolor']
data3=data.loc[data['Classification']=='Iris-virginica']

plt.boxplot([data1['SL'],data2['SL'],data3['SL']])
plt.xticks([1, 2, 3], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.title("Sepal Length Vs Classification ",fontsize=20,pad=20)
plt.xlabel("Classification",fontsize=15,labelpad=10)
plt.ylabel("Sepal Length",fontsize=15,labelpad=10)

#Here we can see Sepal Length ditribution of each species
```

```python
#Plot3 - Joint plot

p=sns.jointplot(x=data.PW,y=data.PL,hue=data['Classification']).set_axis_labels('Petal Width', 'Petal Length')
p.fig.suptitle("Petal length and width distribution of each species",y=1.02, fontsize = 16)

#Here we can see Petal length and width distribution of each species
```
