# Initial setup

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 1. Read the dataset to the python environment.

```python

data1 = pd.read_csv("D:\Data Science and Analytics\CaseStudy 1\cars_data.csv")
data1
```

# 2. Check for the null values present in the dataset. 

```python
print(pd.isnull(data1).sum())
k=0
for i in data1:
    k+=data1[i].isnull().sum()
print("There are",k,"null values present in the dataset.") 
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.countplot(x=data1["Buyer Gender"],ax=ax,data=data1).set_title("Male Vs Female Buyers (Count)",fontsize=20,pad=15)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ages = pd.DataFrame()
bins = [20, 30, 40, 50, 60, 70, 80]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
ages['Age Range'] = pd.cut(data1["Buyer Age"], bins, labels = labels,include_lowest = True)

sns.countplot(x=data1["Buyer Gender"],hue=ages['Age Range'],ax=ax,data=data1).set_title("Male Vs Female Buyers (Age Distribution)",fontsize=20,pad=15)
```
