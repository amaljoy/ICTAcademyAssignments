# Case Study on ANOVA
## XYZ Company has offices in four different zones. The company wishes to investigate the following :
### ● The mean sales generated by each zone.
### ● Total sales generated by all the zones for each month.
### ● Check whether all the zones generate the same amount of sales.
## Help the company to carry out their study with the help of data provided.

## Initial Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st
```

## Reading Data

```python
df = pd.read_csv("D:\Data Science and Analytics\Casestudy 5\Sales_data_zone_wise.csv")
```

```python
df.head()
```

```python
pd.options.display.float_format = '{:.0f}'.format
df.describe()
```

## 1) Find mean sales by each zone

```python
for column in df:
    if(column!="Month"):
        print("Mean sales of",column,"is",df[column].mean())
```

## 2) Find total sales generated by all the zones for each month.

```python
column_list = list(df)
column_list.remove("Month")
column_list.remove("sum")
df["sum"] = df[column_list].sum(axis=1)
for ind in df.index:
     print("Total sales of ",df['Month'][ind],"is", df['sum'][ind])
```

# 3) Check whether all the zones generate the same amount of sales.

## Inorder to check whether all zones generate the same amount of sales, let's do a one-way-anova test
### • Ho - All zones generate same amounts of sales
### • Ha - Zones generate different amounts of sales
### • α - 0.05

```python
fvalue, pvalue = st.f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'], df['Zone - D'])
print("The F value is",fvalue,"\nThe P value is", pvalue)
```

## Insights
### • The p-value 0.00118 is significantly less than considered level of significance 0.05. 
### • So we can reject our Null Hypothesis (Ho) 
### • Zones generate diiferent amounts of sales
